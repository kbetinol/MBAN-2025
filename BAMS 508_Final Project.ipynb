### **(a) Build the Algebraic Formulation**
_a)	Formulate (algebraically) an optimization problem to help the KPD program maximize the number of achievable transplants._

#### **Decision Variables:**
Let $\ x_{ij} $ be a binary decision variable, where:

$$
x_{ij} = 
\begin{cases}
1 & \text{if the donor in pair } i \text{ donates to the patient in pair } j, \\
0 & \text{otherwise}.
\end{cases}
$$

#### **Objective:**
The objective is to maximize the number of transplants (i.e., the total number of pairs where a kidney is exchanged):

$$
\text{Maximize} \ Z = \sum_{(i,j) \in A} x_{ij}
$$

where \( A \) is the set of feasible arcs (i.e., the set of pairs \( (i,j) \) where donor \( i \) can donate to patient \( j \), based on compatibility).

#### **Constraints:**
#### 1) Each patient receives at most one kidney:
For each patient \(j\), the total number of kidneys received by them must be at most 1:

$$
\sum_{i \in P} x_{ij} \leq 1, \quad \forall j \in P
$$

where \( P \) is the set of all patient pairs.

#### 2) Each donor donates at most one kidney:
For each donor \(i\), the total number of kidneys donated by them must be at most 1:

$$
\sum_{j \in P} x_{ij} \leq 1, \quad \forall i \in P
$$

#### 3) Flow balance for cycles (Inflow = Outflow for each pair):
For each pair \(i\), the number of kidneys received by them must be equal to the number of kidneys donated by them:

$$
\sum_{k \in P} x_{ki} = \sum_{j \in P} x_{ij}, \quad \forall i \in P
$$

This constraint ensures the flow of kidneys forms cycles, i.e., if a donor from pair \(i\) donates a kidney to patient \(j\), then the kidney is received back through a matching flow.

#### 4) Compatibility constraints:
This ensures that a donor \(i\) can only donate a kidney to patient \(j\) if the blood and tissue types match. Specifically, for each feasible pair \(i,j\):

- The donor's blood type $\ dBT_i $ must be compatible with the recipient's blood type $\ pBT_j $ based on the compatibility rules.
- The tissue type of the recipient $\ pTT_j $ must match the donor's tissue type $\ dTT_i $.

Mathematically, the compatibility condition is:

$$
\text{if } \ pTT_j = dTT_i \text{ and } pBT_j \in \text{compatibility}(dBT_i), \text{ then } x_{ij} = 1
$$

Otherwise,

$$
x_{ij} = 0
$$

#### **Non-negativity Constraints:**
$$x_{ij} \ge 0 \text{ for all i}$$

### **Sets and Indices**
- $\ (P) $: Set of patient-donor pairs ($\ P = {1, 2, \ldots, n} $).
- $\ (i, j) $: Indices representing patient-donor pairs in the set $\ P $.


### **Parameters**
- $\ C[i][j] $: Compatibility indicator between donor \(i\) and patient \(j\), where:
  $$
  C[i][j] = 
  \begin{cases} 
  1 & \text{if donor \( i \) is compatible with patient \( j \)}, \\
  0 & \text{otherwise}.
  \end{cases}
  $$


### **Decision Variables**
- $\ x[i][j] $: Binary variable indicating whether donor \(i\) donates to patient \(j\), where:
  $$
  x[i][j] =
  \begin{cases}
  1 & \text{if donor \( i \) donates to patient \( j \)}, \\
  0 & \text{otherwise}.
  \end{cases}
  $$


### **Objective Function**
Maximize the total number of transplants:
$$
\text{Maximize } \sum_{i \in P} \sum_{j \in P} x[i][j].
$$


### **Constraints**
1. **Compatibility Constraint**:
   Ensure that a transplant can only occur if the donor and patient are compatible:
  $$
   x[i][j] \leq C[i][j], \quad \forall i, j \in P.
  $$

2. **Reciprocity Constraint**:
   Ensure that the total number of donations from donor \( i \) equals the total number of transplants received by patient \( i \):
  $$
   \sum_{j \in P} x[i][j] = \sum_{j \in P} x[j][i], \quad \forall i \in P.
   $$

3. **Single Donation Constraint**:
   Ensure that each donor \( i \) can donate to at most one patient:
   $$
   \sum_{j \in P} x[i][j] \leq 1, \quad \forall i \in P.
  $$

4. **Single Receipt Constraint**:
   Ensure that each patient \( j \) can receive a transplant from at most one donor:
   $$
   \sum_{i \in P} x[i][j] \leq 1, \quad \forall j \in P.
  $$


### **Summary of the Model**
- **Objective**: Maximize the total number of compatible transplants.
- **Decision Variables**: Binary variables $\ x[i][j] $ indicate whether a transplant occurs between donor \(i\) and patient \(j\).
- **Constraints**: Ensure compatibility, reciprocity, and limit each donor to one donation and each patient to one receipt.

_Solve your formulation from part a) using Python/Gurobi to determine the maximum number of transplants achievable from this pool, along with an indication of which donors should give to which patients to achieve this total._

```python
from gurobipy import Model, GRB, quicksum
import pandas as pd

# Load the data
data = pd.read_csv("KPD_pool.csv")

n = len(data)

# Blood type compatibility rules
compatibility = {
    'A':   ['A','AB'],
    'B':   ['B','AB'],
    'AB':  ['AB'],
    'O':   ['A','B','AB','O']
}

# Create sets
pairs = range(n)  # 0-based indexing internally
# We will refer to data[i] as pair i+1 in output because original data is 1-based.

# Determine feasible arcs
A = []
for i in pairs:
    pBT_i = data.loc[i, 'PatientBloodType']  # Patient Blood Type for pair i
    pTT_i = data.loc[i, 'PatientTissueType']  # Patient Tissue Type for pair i
    dBT_i = data.loc[i, 'DonorBloodType']  # Donor Blood Type for pair i
    dTT_i = data.loc[i, 'DonorTissueType']  # Donor Tissue Type for pair i
    for j in pairs:
        if i != j:
            pBT_j = data.loc[j, 'PatientBloodType']  # Patient Blood Type for pair j
            pTT_j = data.loc[j, 'PatientTissueType']  # Patient Tissue Type for pair j
            # Check if donor i can donate to patient j
            # Condition: pTT_j == dTT_i (tissue type match), and pBT_j in compatibility[dBT_i]
            if pTT_j == dTT_i and pBT_j in compatibility[dBT_i]:
                A.append((i,j))

# Build and solve model
m = Model("Kidney_Exchange")

# Decision variables
x = m.addVars(A, vtype=GRB.BINARY, name="x")

# Objective: maximize sum of x_ij
m.setObjective(quicksum(x[i,j] for (i,j) in A), GRB.MAXIMIZE)

# Constraints
# 1) Each pair receives at most one kidney
for j in pairs:
    m.addConstr(quicksum(x[i,j] for (i,j2) in A if j2 == j) <= 1, name=f"Receive_{j+1}")

# 2) Each pair donates at most one kidney
for i in pairs:
    m.addConstr(quicksum(x[i,j] for (i2,j) in A if i2 == i) <= 1, name=f"Donate_{i+1}")

# 3) Flow balance for cycles: For each pair i, Inflow = Outflow
for i in pairs:
    inflow = quicksum(x[k,i] for (k,i2) in A if i2 == i)
    outflow = quicksum(x[i,j] for (i2,j) in A if i2 == i)
    m.addConstr(inflow - outflow == 0, name=f"Flow_{i+1}")

# Add Gurobi parameters to find all optimal solutions
m.setParam(GRB.Param.PoolSolutions, 1000)  # Set a high pool limit
m.setParam(GRB.Param.PoolSearchMode, 2)    # Enumerate all optimal solutions
m.optimize()

# Output all optimal solutions
if m.Status == GRB.OPTIMAL:
    solution_count = m.SolCount
    
    # Fetch the objective value of the optimal solution
    m.setParam(GRB.Param.SolutionNumber, 0)
    optimal_value = m.PoolObjVal
    print(f"Optimal objective value: {optimal_value}\n")
    
    # Iterate through all solutions
    for sol_num in range(solution_count):
        m.setParam(GRB.Param.SolutionNumber, sol_num)  # Access each solution
        if m.PoolObjVal == optimal_value:  # Check if it matches the optimal value
            matches = []
            
            # Extract donor-recipient matches for the current solution
            for (i, j) in A:
                if x[i, j].Xn > 0.5:  # Using Xn to fetch value from current solution
                    matches.append((i + 1, j + 1))  # 1-based indexing
            
            # Print the solution details
            if matches:
                print(f"Optimal Solution #{sol_num + 1}:")
                for match in matches:
                    print(f"Donor from pair {match[0]} donates to patient from pair {match[1]}")
                
                print(f"Total number of transplants: {len(matches)}\n")
else:
    print("No optimal solution found.")
```

### **Algebraic Formulation**

Let:

- $\ x_{ij}^{orig}$ be a binary decision variable where:
  - $\ x_{ij}^{orig} = 1$ if donor \(i\) donates to patient \(j\) in the original (direct) match,
  - $\ x_{ij}^{orig} = 0$ otherwise.

- $\ x_{ij}^{conv}$ be a binary decision variable where:
  - $\ x_{ij}^{conv} = 1$ if donor \(i\) donates to patient \(j\) in the conversion match (blood type O donor),
  - $\ x_{ij}^{conv} = 0$ otherwise.

- $\ y_i $ be a binary decision variable where:
  - $\ y_i = 1$ if donor \(i\) is used in a conversion match,
  - $\ y_i = 0$ otherwise.

#### **Objective Function:**

Maximize the total number of transplants (original and conversion matches):

$$
\text{Maximize} \ Z = \sum_{(i,j) \in \text{possible\_orig}} x_{ij}^{orig} + \sum_{(i,j) \in \text{possible\_conv}} x_{ij}^{conv}
$$

#### **Constraints:**

1. **Conversion Limit**: The number of conversion matches is limited by a constant \( C \):

$$
\sum_{i \in P} y_i \leq C
$$

2. **No Multiple Kidney Donations**: Each donor can donate at most one kidney:

$$
\sum_{j \in P} x_{ij}^{orig} + \sum_{j \in P} x_{ij}^{conv} \leq 1, \quad \forall i \in P
$$

3. **No Multiple Kidney Reception**: Each patient can receive at most one kidney:

$$
\sum_{i \in P} x_{ij}^{orig} + \sum_{i \in P} x_{ij}^{conv} \leq 1, \quad \forall j \in P
$$

4. **Conversion Constraints**: Conversion match can only occur if the original match is not used:

$$
x_{ij}^{conv} \leq 1 - x_{ij}^{orig}, \quad \forall (i,j) \in \text{possible\_orig}
$$

5. **Linking Constraints**: A conversion match can only happen if the donor is used in a conversion:

$$
x_{ij}^{orig} \leq 1 - y_i, \quad \forall (i,j) \in \text{possible\_orig}
$$

$$
x_{ij}^{conv} \leq y_i, \quad \forall (i,j) \in \text{possible\_conv}
$$

6. **Flow Balance**: For each donor-recipient pair, the kidneys donated and received must balance (for kidney exchange cycles):

$$
\sum_{k \in P} x_{ki}^{orig} + \sum_{k \in P} x_{ki}^{conv} = \sum_{j \in P} x_{ij}^{orig} + \sum_{j \in P} x_{ij}^{conv}, \quad \forall i \in P
$$

### **Python Code Using Gurobi**

```python
import matplotlib.pyplot as plt
from gurobipy import Model, GRB, quicksum
import pandas as pd

# Load the data
data = pd.read_csv("KPD_pool.csv")

n = len(data)

compatibility = {
    'A':   ['A','AB'],
    'B':   ['B','AB'],
    'AB':  ['AB'],
    'O':   ['A','B','AB','O']
}

pairs = range(n)

def feasible_arcs(donorBT, donorTT):
    def can_donate(patientBT, patientTT):
        return (patientTT == donorTT) and (patientBT in compatibility[donorBT])
    return can_donate

# Precompute feasibility for original and converted (O) scenarios
possible_orig = {}
possible_conv = {}
for i in pairs:
    pBT_i = data.loc[i, 'PatientBloodType']  # Patient Blood Type for pair i
    pTT_i = data.loc[i, 'PatientTissueType']  # Patient Tissue Type for pair i
    dBT_i = data.loc[i, 'DonorBloodType']  # Donor Blood Type for pair i
    dTT_i = data.loc[i, 'DonorTissueType']  # Donor Tissue Type for pair i
    can_donate_orig = feasible_arcs(dBT_i, dTT_i)
    can_donate_O = feasible_arcs('O', dTT_i)

    for j in pairs:
        if i == j:
            continue

        pBT_j = data.loc[j, 'PatientBloodType']  # Patient Blood Type for pair j
        pTT_j = data.loc[j, 'PatientTissueType']  # Patient Tissue Type for pair j

        # Determine if the pair is feasible for original match
        is_orig_possible = can_donate_orig(pBT_j, pTT_j)
        if is_orig_possible:
            possible_orig[(i, j)] = True  # This pair is feasible for original match

        # Only add to possible_conv if not already valid for original match
        if not is_orig_possible and can_donate_O(pBT_j, pTT_j):
            possible_conv[(i, j)] = True  # This pair is feasible for conversion match
print("Possible Original Matches:", possible_orig)
print("Possible Conversion Matches:", possible_conv)
        
def print_matches(x_orig, x_conv):
    original_matches = []
    conversion_matches = []

    for (i, j) in x_orig:
        if x_orig[(i, j)].x > 0.5:  # if x_orig is 1 (match made)
            original_matches.append((i+1, j+1))  # store 1-based index

    for (i, j) in x_conv:
        if x_conv[(i, j)].x > 0.5:  # if x_conv is 1 (conversion made)
            conversion_matches.append((i+1, j+1))  # store 1-based index

    print("Original Matches:")
    for match in original_matches:
        print(f"Pair {match[0]} (Donor) -> Pair {match[1]} (Recipient)")

    print("\nConversion Matches:")
    for match in conversion_matches:
        print(f"Pair {match[0]} (Donor) -> Pair {match[1]} (Recipient)")


def solve_with_conversions(C):
    """
    Solve the kidney exchange problem allowing up to C conversions.
    Returns the objective value and solution details.
    """
    m = Model("Kidney_Exchange_Conversions")
    m.setParam('OutputFlag', 0)

    # Variables
    y = m.addVars(pairs, vtype=GRB.BINARY, name="y")

    x_orig = {}
    x_conv = {}
    
# Iterate over all unique pairs
    for (i, j) in possible_orig:
        # Check if the pair is in possible_orig before accessing
        if possible_orig[(i, j)]:
            x_orig[(i, j)] = m.addVar(vtype=GRB.BINARY, name=f"x_orig_{i+1}_{j+1}")

    for (i, j) in possible_conv:
    # Check if the pair is in possible_conv before accessing
        if (i, j) in possible_conv and possible_conv[(i, j)]:
            x_conv[(i, j)] = m.addVar(vtype=GRB.BINARY, name=f"x_conv_{i+1}_{j+1}")
    
    # Objective: Maximize total transplants (both original and conversion matches)
    obj_terms = []
    
    # Summing over possible original pairs
    for (i, j) in possible_orig:
        if (i, j) in x_orig:  # Ensure the pair is in x_orig before adding
            obj_terms.append(x_orig[(i, j)])

    # Summing over possible conversion pairs
    for (i, j) in possible_conv:
        if (i, j) in x_conv:  # Ensure the pair is in x_conv before adding
            obj_terms.append(x_conv[(i, j)])

    m.setObjective(quicksum(obj_terms), GRB.MAXIMIZE)

    # Constraints: Conversion limit
    m.addConstr(quicksum(y[i] for i in pairs) <= C, name="Conversion_Limit")
    
    # Constraints to ensure no one gets multiple kidneys (receiving and donating)
    for i in pairs:
        for j in pairs:
            if i == j: continue
            if (i, j) in x_conv and (i, j) in x_orig:
                # Conversion can only happen if the original match is not used
                m.addConstr(x_conv[(i, j)] <= 1 - x_orig[(i, j)], name=f"Priority_orig_{i+1}_{j+1}")
            if (i, j) in x_orig:
                m.addConstr(x_orig[(i, j)] <= 1 - y[i], name=f"Link_orig_{i+1}_{j+1}")
            if (i, j) in x_conv:
                m.addConstr(x_conv[(i, j)] <= y[i], name=f"Link_conv_{i+1}_{j+1}")

    # Receive at most one kidney
    for j in pairs:
        incoming_arcs = []
        for i in pairs:
            if i == j: continue
            if (i, j) in x_orig:
                incoming_arcs.append(x_orig[(i, j)])
            if (i, j) in x_conv:
                incoming_arcs.append(x_conv[(i, j)])
        m.addConstr(quicksum(incoming_arcs) <= 1, name=f"Receive_{j+1}")

    # Donate at most one kidney
    for i in pairs:
        outgoing_arcs = []
        for j in pairs:
            if i == j: continue
            if (i, j) in x_orig:
                outgoing_arcs.append(x_orig[(i, j)])
            if (i, j) in x_conv:
                outgoing_arcs.append(x_conv[(i, j)])
        m.addConstr(quicksum(outgoing_arcs) <= 1, name=f"Donate_{i+1}")

    # Flow balance
    for i in pairs:
        incoming_arcs = []
        outgoing_arcs = []
        for k in pairs:
            if k == i: continue
            if (k, i) in x_orig:
                incoming_arcs.append(x_orig[(k, i)])
            if (k, i) in x_conv:
                incoming_arcs.append(x_conv[(k, i)])
            if (i, k) in x_orig:
                outgoing_arcs.append(x_orig[(i, k)])
            if (i, k) in x_conv:
                outgoing_arcs.append(x_conv[(i, k)])
        m.addConstr(quicksum(incoming_arcs) - quicksum(outgoing_arcs) == 0, name=f"Flow_balance_{i+1}")

    m.optimize()

    # Return the objective value if the optimization is successful
    if m.status == GRB.OPTIMAL:
        print("Optimal Solution Found")
        print(f"Total Objective Value: {m.objVal}")
        print_matches(x_orig, x_conv)
        return m.objVal  # Return the objective value
    else:
        print("No optimal solution found.")
        return 0  # Return 0 if no optimal solution is found

# Call the function to solve with a conversion limit, e.g., C = 2
#solve_with_conversions(2)

# Collect results for different values of C
results = []
for C in range(0, n+1):
    val = solve_with_conversions(C)
    results.append((C, val))

# Plot results
Cs = [r[0] for r in results]
Vals = [r[1] for r in results]

plt.figure(figsize=(8, 5))
plt.plot(Cs, Vals, marker='o')
plt.xlabel('Number of Blood Conversions (C)')
plt.ylabel('Total Transplants')
plt.title('Total Transplants vs. Number of Donor Blood Type Conversions')
plt.grid(True)
plt.show()
```

> - At the start (0–6 conversions), the number of transplants increases steeply. This suggests that converting even a small number of donor kidneys to type O significantly improves compatibility and enables more matches.
> - This steep rise is likely due to the ability of type O kidneys to be universally accepted by recipients with any blood type.  

> #### **2. Plateauing Effect:**
> - After approximately 6 donor blood type conversions, the total number of transplants stabilizes at a maximum of 20 transplants.
> - This implies that once critical compatibility bottlenecks are resolved, additional conversions provide diminishing returns. The system reaches its full capacity for transplants given the current pool of donors and patients.
